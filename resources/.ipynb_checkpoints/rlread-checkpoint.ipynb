```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
trueQa = np.random.randn(10,1)
print(trueQa)
```

```python
np.argmax(trueQa)
```

```python
def nArmBandit(qa,ep,T=1000):
    allq = 5*np.ones((10,1))
    alln = np.zeros((10,1))
    sumr = 0
    pltavgr = [0]
    for i in range(1,T):
        greedy = (np.random.rand(1)<=1-ep)
        if greedy:
            a = np.argmax(allq)
        else:
            a = np.random.choice(10)
        alln[a] += 1
        r = trueQa[a]+np.random.randn(1)
        allq[a] += r/alln[a]
        sumr += r
        pltavgr.append(sumr/i)
    return pltavgr
```

```python
sample1 = nArmBandit(trueQa,0)
sample2 = nArmBandit(trueQa,0.01)
sample3 = nArmBandit(trueQa,0.1)
```

```python
plt.plot(sample1)
plt.plot(sample2)
plt.plot(sample3)
plt.legend(['1','2','3'])
```
