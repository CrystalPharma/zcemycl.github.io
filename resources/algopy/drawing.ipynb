```python
import numpy as np
import math
import networkx as nx
from nxpd import draw
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_edges_from([('a b c d e f','a b c'),
                  ('a b c d e f','d e f')])
draw(G, show='ipynb')
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_edges_from([('a b \n c d e f','a b'),
                  ('a b c d e f','c d'),
                  ('a b c d e f','e f')])
draw(G, show='ipynb')
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_edges_from([('[a b c d]','[a b]'),
                  ('[a b c d]','[c d]'),
                  ('[a b]','[a]'),
                  ('[a b]','[b]'),
                  ('[c d]','[c]'),
                  ('[c d]','[d]')])
draw(G, show='ipynb')
```
