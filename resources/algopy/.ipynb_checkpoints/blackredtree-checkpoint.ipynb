```python
G = nx.DiGraph()
G.add_edge(30,20)
G.add_edge(30,40)
G.add_edge(40,'alpha')
G.add_edge(40,50)
draw(G, show='ipynb')
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_node(3,pos=(2,3))
G.add_node(1,pos=(1,2))
G.add_node(5,pos=(3,2))
G.add_node(2,pos=(1.5,1))
G.add_node(4,pos=(2.5,1))
G.add_edges_from([(3,1),(3,5),(1,2),(5,4)])
pos=nx.get_node_attributes(G,'pos')
print(pos)
color_map = ['red','white','white','white','white']
labels = {}
for node in G:
#     color_map.append('white')
    labels[node] = r"$\alpha$"
nx.draw(G,pos,labels=labels,node_color=color_map,
        with_labels=True,linewidths=2)
ax = plt.gca() # to get the current axis
ax.collections[0].set_edgecolor("black") 
plt.axis("off")
plt.show()
```
