```python
import numpy as np
import math
import networkx as nx
from nxpd import draw
import matplotlib.pyplot as plt
```

```python
lst = [['a','b',4],['a','h',8],['b','h',11],
      ['b','c',8],['h','i',7],['i','c',2],['c','f',4],
      ['g','f',2],['h','g',1],['i','g',6],['d','f',14],
      ['d','e',9],['f','e',10],['c','d',7]]
```

```python
G = nx.Graph()
G.graph['rankdir']='LR'
# G.add_node('a',color='black',position=(0,1))
# G.add_node('h',color='black',position=(1,0))
# G.add_node('b',color='black',position=(1,2))
for idx,item in enumerate(lst):
    G.add_edge(item[0],item[1],weight=item[2],
              label=str(item[2]))
#     if idx==2: break
draw(G,show='ipynb')
```
