```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_edges_from([('a b c d e f','a b c'),
                  ('a b c d e f','d e f')])
draw(G, show='ipynb')
```

```python
G=nx.Graph()
G.add_node(1,pos=(1,1))
G.add_node(2,pos=(2,2))
G.add_edge(1,2)
pos=nx.get_node_attributes(G,'pos')
nx.draw(G,pos)
print(pos)
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_node(3,pos=(2,3))
G.add_node(1,pos=(1,2))
G.add_node(5,pos=(3,2))
G.add_node(2,pos=(1.5,1))
G.add_node(4,pos=(2.5,1))
G.add_edges_from([(3,1),(3,5),(1,2),(5,4)])
pos=nx.get_node_attributes(G,'pos')
print(pos)
color_map = []
labels = {}
for node in G:
    color_map.append('white')
    labels[node] = r"$a$"
nx.draw(G,pos,node_color=color_map, with_labels=True,linewidths=2)
nx.draw_networkx_edge_labels(G,pos,edge_labels={(3,1):'less',\
(3,5):'greater',(1,2):'greater',(5,4):'less'},font_color='black')
ax = plt.gca() # to get the current axis
ax.collections[0].set_edgecolor("black") 
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_node(3,pos=(2,3))
G.add_node(1,pos=(1,2))
G.add_node(5,pos=(3,2))
G.add_node(2,pos=(1.5,1))
G.add_node(4,pos=(2.5,1))
G.add_edges_from([(3,1),(3,5),(1,2),(5,4)])
pos=nx.get_node_attributes(G,'pos')
print(pos)
color_map = []
labels = {}
for node in G:
    color_map.append('white')
    labels[node] = r"$a$"
nx.draw(G,pos,labels=labels,node_color=color_map,
        with_labels=True,linewidths=2)
# nx.draw_networkx_labels(G, pos, labels, font_size=16)
nx.draw_networkx_edge_labels(G,pos,edge_labels={(3,1):'less',\
(3,5):'greater',(1,2):'greater',(5,4):'less'},font_color='red')
ax = plt.gca() # to get the current axis
ax.collections[0].set_edgecolor("black") 
plt.axis("off")
plt.show()
```

```python

G = nx.cubical_graph()
pos = nx.spring_layout(G)  # positions for all nodes

# nodes
options = {"node_size": 500, "alpha": 0.8}
nx.draw_networkx_nodes(G, pos, nodelist=[0, 1, 2, 3], node_color="r", **options)
nx.draw_networkx_nodes(G, pos, nodelist=[4, 5, 6, 7], node_color="b", **options)

# edges
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[(0, 1), (1, 2), (2, 3), (3, 0)],
    width=8,
    alpha=0.5,
    edge_color="r",
)
nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[(4, 5), (5, 6), (6, 7), (7, 4)],
    width=8,
    alpha=0.5,
    edge_color="b",
)


# some math labels
labels = {}
labels[0] = r"$a$"
labels[1] = r"$b$"
labels[2] = r"$c$"
labels[3] = r"$d$"
labels[4] = r"$\alpha$"
labels[5] = r"$b$"
labels[6] = r"$\gamma$"
labels[7] = r"$\delta$"
nx.draw_networkx_labels(G, pos, labels, font_size=16)

plt.axis("off")
plt.show()
```
