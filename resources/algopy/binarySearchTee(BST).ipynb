```python
import numpy as np
import math
import networkx as nx
from nxpd import draw
```

```python
class Node():
    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key
```

```python
def insert(root,key):
    if root is None:
        return Node(key)
    else:
        if root.val == key:
            return root
        elif root.val < key:
            root.right = insert(root.right,key)
        elif root.val > key:
            root.left = insert(root.left,key)
    return root
```

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val)
        inorder(root.right)
```

```python
r = Node(50) 
r = insert(r, 30) 
r = insert(r, 20) 
r = insert(r, 40) 
r = insert(r, 70) 
r = insert(r, 60) 
r = insert(r, 80) 
```

```python
inorder(r)
```

```python
# search for the branch starting with the node of 
# key
def search(root,key):
    if root is None or root.val == key:
        return root
    if root.val < key:
        return search(root.right,key)
    elif root.val > key:
        return search(root.left,key)
```

```python
r_ = search(r,80)
inorder(r_)
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
G.add_nodes_from([20,30,40,50,60,70,80])
G.add_edges_from([(50,30),(50,70),(30,20),
                  (30,40),(70,60),(70,80)])
draw(G, show='ipynb')
```

```python
def invertTree(root):
    if root is None:
        return root
    root.left,root.right = root.right,root.left
    invertTree(root.left)
    invertTree(root.right)
    return root
```

```python
r = invertTree(r)
inorder(r)
```

```python
G = nx.DiGraph()
G.graph['dpi'] = 120
stack = [r]
while True:
    tmp = stack.pop(0)
    if tmp.left is not None:
        G.add_edges_from([(tmp.val,tmp.left.val)])
        stack.append(tmp.left)
    if tmp.right is not None:
        G.add_edges_from([(tmp.val,tmp.right.val)])
        stack.append(tmp.right)
    
    if not stack:
        break
draw(G, show='ipynb')
```
